```python
class TrafficDevice:
    def activate(self):
        print("Traffic Devices have been activated")


class Trafficlight(TrafficDevice):
    def activate(self):
        print("Traffic light has been activated")

class SpeedCamera(TrafficDevice):
    def activate(self):
        print("Speed Camera has been activated")

class PedestrianSignal(TrafficDevice):
    def activate(self):
        print("Pedestrian Signal is now activated")

class EmergencySiren(TrafficDevice):
    def activate(self):
        print("Emergency Siren is sounding")
```

```python
traffic_devices = [Trafficlight(),SpeedCamera(),PedestrianSignal(),EmergencySiren()]
```

```python
for dev in traffic_devices:
    dev.activate()
```
